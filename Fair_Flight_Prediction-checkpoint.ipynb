```python
#it is the difference between Departure time and Arrival time 
duration = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
from sklearn.ensemble import ExtraTreesRegressor

# Initialize the model
selection = ExtraTreesRegressor()

# Fit the model with your data
selection.fit(X, y)

# Now you can access the feature importances
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)

# Plot the top 20 important features
plt.figure(figsize=(12, 8))
feat_importances.nlargest(20).plot(kind='barh')
plt.title('Top 20 Feature Importances')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure selection is a trained model and X is your feature DataFrame
plt.figure(figsize=(12, 8))

# Convert feature importances into a Pandas Series for better visualization
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)

# Select the top 20 features by importance and plot them as a horizontal bar graph
feat_importances.nlargest(20).plot(kind='barh')

plt.title('Top 20 Feature Importances')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.show()
```
